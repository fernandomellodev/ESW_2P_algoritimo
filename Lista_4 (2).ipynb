```python
"""1. Escreva uma função recursiva que calcule a soma dos dígitos de um número inteiro positivo.
Exemplo: 123 =>6"""

def soma_digitos (n):
    if n < 10:
        return n
    return (n % 10) + soma_digitos(n // 10)
print (f"soma dos digitos de 123: {soma_digitos(123)}")
```

```python
"""2. Crie uma função recursiva que conte quantos dígitos tem um número inteiro.
Exemplo: 9876 =>4"""

def contar_digitos(n):
    if n < 10:
        return 1
    return 1 + contar_digitos(n // 10)

print (f"Digitos em 9878: {contar_digitos(9876)}")
print (f"Digitos em 5: {contar_digitos(5)}")
```

```python
"""5. Crie uma função recursiva que calcule o produto de dois números usando apenas soma.
Exemplo: 3 × 4 =>12"""

def produto_soma(a, b):
    if b == 0:
        return 0
    return a + produto_soma(a, b -1)
print(f"Produto de 3 x 4: {produto_soma(3, 4)}")
print(f"Produto de 5 x 2: {produto_soma(5, 2)}")
```

```python
"""6. Implemente uma função recursiva que encontre o maior elemento de uma lista.
"""
def maior_elemento(lista):
    if len(lista) == 1:
        return lista[0]
    maior_resto = maior_elemento(lista[1:])

    if lista [0] > maior_resto:
        return lista[0]
    return maior_resto

print(f"Maior de [1, 5, 3]: {maior_elemento([1, 5, 3])}") 
print(f"Maior de [10, 2, 45, 8]: {maior_elemento([10, 2, 45, 8])}")
```

```python
"""8. Crie uma função recursiva que conte quantas vezes um elemento aparece em uma lista.
"""
def contar_elemento(lista, item):
    if not lista:
        return 0
    aparece = 1 if lista[0] == item else 0
    return aparece + contar_elemento(lista[1:], item)
minha_lista = [1, 2, 1, 3, 1]
print(f"O número 1 aparece {contar_elemento(minha_lista, 1)} vezes.") 
print(f"O número 5 aparece {contar_elemento(minha_lista, 5)} vezes.")
```
