```python

"""Capturando exceções múltiplas: Crie um programa que peça ao usuário o
nome de uma cor e mostre seu valor em RGB de acordo com um dicionário
pré-definido. O programa deve tratar exceções caso o nome da cor não exista
no dicionário.
cores = {'vermelho': (255, 0, 0), 'verde': (0, 255, 0), 'azul':
(0, 0, 255)}"""



cores = {
    'vermelho': (255, 0, 0), 
    'verde': (0, 255, 0), 
    'azul': (0, 0, 255)
}

try:
    nome_cor = input("Digite o nome de uma cor (vermelho, verde ou azul): ").strip().lower()
    
    rgb = cores[nome_cor]
    
    print(f"O valor RGB de '{nome_cor}' é: {rgb}")

except KeyError:
    # Este bloco captura especificamente quando a chave não é encontrada
    print(f"Erro: A cor '{nome_cor}' não está cadastrada no nosso dicionário.")

except KeyboardInterrupt:
    # Captura se o usuário pressionar Ctrl+C para encerrar o programa
    print("\nBusca cancelada pelo usuário.")

except Exception as e:
    # Captura qualquer outro erro imprevisto
    print(f"Ocorreu um erro inesperado: {e}")

finally:
    print("Obrigado por usar o consultor de cores!")
    
    
    
```

```python
"""Bloco else e finally: Escreva um programa que solicite um número ao
usuário. Se o número for maior que 10, exiba uma mensagem dizendo que o
número é válido. Utilize o bloco else para imprimir que o programa foi
executado com sucesso, e o bloco finally para imprimir "Programa
encerrado"."""

try:
    # Solicita o número e converte para float para aceitar decimais
    numero = float(input("Digite um número: "))
    
    if numero > 10:
        print(f"O número {numero} é válido!")
    else:
        print(f"O número {numero} não é maior que 10.")

except ValueError:
    # Caso o usuário digite texto em vez de números
    print("Erro: Entrada inválida! Você deve digitar um número.")

else:
    # Este bloco só roda se o 'try' foi executado sem erros
    print("O programa foi executado com sucesso (sem erros de entrada).")

finally:
    # Este bloco sempre roda, aconteça o que acontecer
    print("Programa encerrado.")
```

```python
""". Exceções personalizadas: Escreva uma função que verifica se uma senha
possui no mínimo 8 caracteres e pelo menos um número. Se a senha não
atender aos requisitos, levante uma exceção com uma mensagem
personalizada. Trate a exceção e mostre a mensagem ao usuário."""

# Definimos nossa própria classe de erro
class SenhaInvalidaError(Exception):
    """Exceção levantada para erros na validação da senha."""
    pass

def verificar_senha(senha):
    # Regra 1: Mínimo de 8 caracteres
    if len(senha) < 8:
        raise SenhaInvalidaError("A senha deve ter pelo menos 8 caracteres.")
    
    # Regra 2: Pelo menos um número
    # O método any() verifica se pelo menos um caractere é dígito
    if not any(char.isdigit() for char in senha):
        raise SenhaInvalidaError("A senha deve conter pelo menos um número.")
    
    return True

# --- Testando a Função ---
try:
    minha_senha = input("Digite sua nova senha: ")
    if verificar_senha(minha_senha):
        print("Senha cadastrada com sucesso!")

except SenhaInvalidaError as e:
    # Aqui capturamos especificamente o nosso erro personalizado
    print(f"Erro de Validação: {e}")

except Exception as e:
    # Segurança para outros erros inesperados
    print(f"Ocorreu um erro inesperado: {e}")
```

```python
"""Simulação de transações: Crie um programa que simule uma transferência
bancária. Peça ao usuário o saldo da conta e o valor da transferência. Caso o
saldo seja insuficiente, levante uma exceção do tipo ValueError com a
mensagem "Saldo insuficiente". Trate a exceção adequadamente e informe o
usuário."""

try:
    # 1. Entrada de dados
    saldo_atual = float(input("Digite o seu saldo atual: R$ "))
    valor_transferencia = float(input("Digite o valor da transferência: R$ "))

    # 2. Regra de Negócio: Verificação de Saldo
    if valor_transferencia > saldo_atual:
        # Levantamos o erro manualmente se a condição falhar
        raise ValueError("Saldo insuficiente para realizar esta operação.")
    
    if valor_transferencia <= 0:
        raise ValueError("O valor da transferência deve ser maior que zero.")

    # 3. Processamento (Se não houve erro acima)
    novo_saldo = saldo_atual - valor_transferencia
    print(f"\n✅ Transferência de R$ {valor_transferencia:.2f} realizada com sucesso!")

except ValueError as e:
    # O 'e' captura a mensagem que escrevemos dentro do raise ValueError(...)
    print(f"\n❌ Erro na transação: {e}")

except Exception as erro_inesperado:
    print(f"\n⚠️ Ocorreu um erro sistêmico: {erro_inesperado}")

else:
    # Executa apenas se a transferência ocorreu sem erros
    print(f"Seu novo saldo é: R$ {novo_saldo:.2f}")

finally:
    print("\nObrigado por utilizar nossos serviços bancários. Tenha um ótimo dia!")
```
