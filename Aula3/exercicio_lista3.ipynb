PARTE 1 – Busca Sequencial
(Percorre o vetor do início ao fim – O(n))

```python
"""1. Básico 

 Escreva uma função busca_sequencial(lista, alvo) que retorna o índice da primeira ocorrência
do alvo na lista. Se não encontrar, retorne -1."""

def busca_sequencial(lista, alvo):

    for i in range(len(lista)):
        if lista[i] == alvo:
            return i

    return -1


dados = [4, 8, 2, 10, 8]

print(busca_sequencial(dados, 10))
```

```python
"""2. Contagem de ocorrências
Modifique a função anterior para retornar quantas vezes o alvo aparece na lista (retorne 0 se
não existir)."""

def contar_ocorrencias(lista, alvo):
    contador = 0
    for elemento in lista:
        if elemento == alvo:
            contador += 1
    return contador

numeros = [1, 2, 5, 3, 7, 9]
print(contar_ocorrencias(numeros, 1))
```

```python
"""3. Menor índice
Dada uma lista de números, encontre o índice do menor valor usando apenas busca
sequencial (sem usar min())."""

def indice_do_menor(lista):
    if not lista:
        return -1
    
    indice_menor = 0
    for i in range(1, len(lista)):
        if lista[i] < lista[indice_menor]:
            indice_menor = i
            
    return indice_menor
```

```python
"""4. Busca em lista de strings
Crie uma função que recebe uma lista de nomes e um nome alvo. Retorne o índice (caseinsensitive).
"""
def busca_nomes_case_insensitive(lista, alvo):
    alvo = alvo.lower()
    for i in range(len(lista)):
        if lista[i].lower() == alvo:
            return i
    return -1
nomes = ["Alice", "Bob", "Charlie", "Diana"]
print(busca_nomes_case_insensitive(nomes, "charlie"))
```

```python
"""5. Busca com condição
Em uma lista de dicionários [{"nome": "João", "idade": 25}, ...], encontre o primeiro aluno com
idade ≥ 18 e retorne o nome."""

def busca_maior_idade(lista):
    for aluno in lista:
        if aluno["idade"] >= 18:
            return aluno["nome"]
    return None

# Exemplo de uso
alunos = [
    {"nome": "Maria", "idade": 16},
    {"nome": "João", "idade": 25},
    {"nome": "Ana", "idade": 17}
]

print(busca_maior_idade(alunos))
```

```python
"""6. Busca em matriz (2D)
Dada uma matriz 3x3, use busca sequencial para encontrar se um número existe e retorne a
posição (linha, coluna) ou (-1, -1)."""

def busca_matriz(matriz, alvo):
    for i in range(len(matriz)):
        for j in range(len(matriz[i])):
            if matriz[i][j] == alvo:
                return (i, j)
    return (-1, -1)

matriz = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(busca_matriz(matriz, 5))
```

```python
"""7. Contar elementos menores que X
Usando busca binária, retorne quantos elementos são menores que o alvo (sem percorrer
toda a lista)."""

import random

lista = [random.randint(1, 100) for _ in range(100)]

def buscar_limite(lista):
    for i, valor in enumerate(lista):
        if valor > 50:
            return i, valor
    return -1, -1

indice, valor = buscar_limite(lista)
print(f"Índice: {indice}, Valor: {valor}")
```

```python
"""8. Busca em matriz ordenada
Dada uma matriz 4x4 onde cada linha e cada coluna está ordenada, encontre se um número
existe (usando busca binária otimizada)."""

def busca_binaria_matriz(matriz, alvo):
    for linha in matriz:
        # Aplica busca binária em cada linha
        inicio, fim = 0, len(linha) - 1
        while inicio <= fim:
            meio = (inicio + fim) // 2
            if linha[meio] == alvo:
                return (matriz.index(linha), meio)
            elif linha[meio] < alvo:
                inicio = meio + 1
            else:
                fim = meio - 1
    return (-1, -1)
```

```python
"""9. Busca com múltiplos critérios
 Em uma lista de tuplas (nome, nota), encontre o primeiro aluno com nota ≥ 7 e nome
começando com “A”. Retorne o nome ou “Não encontrado”."""

alunos = [
    ("Pedro", 8.5),
    ("Ana", 6.0),
    ("Arthur", 7.5),
    ("Beatriz", 9.0)
]

def busca_criterios(lista):
    for nome, nota in lista:
        if nota >= 7 and nome.startswith("A"):
            return nome
    return "Não encontrado"

print(busca_criterios(alunos))
```

PARTE 2 – Busca Binária
(Lista DEVE estar ordenada – O(log n))

```python
"""
 Escreva busca_binaria(lista_ordenada, alvo) que retorna o índice ou -1."""
def busca_binaria(lista_ordenada, alvo):
    inicio, fim = 0, len(lista_ordenada) - 1
    while inicio <= fim:
        meio = (inicio + fim) // 2
        if lista_ordenada[meio] == alvo:
            return meio
        elif lista_ordenada[meio] < alvo:
            inicio = meio + 1
        else:
            fim = meio - 1
    return -1
print(busca_binaria([1, 3, 5, 7, 9], 5))
```

```python
""" Menor índice (primeira ocorrência) 
 Em uma lista com duplicatas (ex: [1, 2, 2, 2, 3]), retorne o menor índice do alvo."""
def busca_primeira_ocorrencia(lista, alvo):
    inicio, fim = 0, len(lista) - 1
    resultado = -1
    while inicio <= fim:
        meio = (inicio + fim) // 2
        if lista[meio] == alvo:
            resultado = meio
            fim = meio - 1  # Continua buscando à esquerda
        elif lista[meio] < alvo:
            inicio = meio + 1
        else:
            fim = meio - 1
    return resultado
print(busca_primeira_ocorrencia([1, 2, 2, 2, 3], 2))
```

```python
"""maior índice (última ocorrência)
 Retorne o maior índice do alvo em lista com duplicatas."""
def busca_ultima_ocorrencia(lista, alvo):
    inicio, fim = 0, len(lista) - 1
    resultado = -1
    while inicio <= fim:
        meio = (inicio + fim) // 2
        if lista[meio] == alvo:
            resultado = meio
            inicio = meio + 1  # Continua buscando à direita
        elif lista[meio] < alvo:
            inicio = meio + 1
        else:
            fim = meio - 1
    return resultado
print(busca_ultima_ocorrencia([1, 2, 2, 2, 3], 2))
```

```python
""" Busca de inserção
 Retorne o índice onde o alvo deveria ser inserido para manter a lista ordenada (mesmo se não
existir)."""
def busca_insercao(lista, alvo):
    inicio, fim = 0, len(lista)
    while inicio < fim:
        meio = (inicio + fim) // 2
        if lista[meio] < alvo:
            inicio = meio + 1
        else:
            fim = meio
    return inicio
print(busca_insercao([1, 3, 5, 7], 4))
```

```python
"""Busca em lista de strings ordenada
 Liste de nomes em ordem alfabética. Encontre um nome (case-insensitive)."""
def busca_nome_ordenado(lista, alvo):
    alvo = alvo.lower()
    inicio, fim = 0, len(lista) - 1
    while inicio <= fim:
        meio = (inicio + fim) // 2
        nome_meio = lista[meio].lower()
        if nome_meio == alvo:
            return meio
        elif nome_meio < alvo:
            inicio = meio + 1
        else:
            fim = meio - 1
    return -1
nomes = ["Alice", "Bob", "Charlie", "Diana"]
print(busca_nome_ordenado(nomes, "charlie"))
```

```python
""" encontrar valor mais próximo
 Se o alvo não existir, retorne o índice do número mais próximo dele na lista ordenada."""
def busca_proximo(lista, alvo):
    inicio, fim = 0, len(lista) - 1
    while inicio <= fim:
        meio = (inicio + fim) // 2
        if lista[meio] == alvo:
            return meio
        elif lista[meio] < alvo:
            inicio = meio + 1
        else:
            fim = meio - 1
    
    if inicio >= len(lista):
        return fim  
    if fim < 0:
        return inicio  
    
    
    if abs(lista[inicio] - alvo) < abs(lista[fim] - alvo):
        return inicio
    else:
        return fim
    

print(busca_proximo([1, 3, 5, 7], 4))
```

```python
"""Contar elementos menores que X
 Usando busca binária, retorne quantos elementos são menores que o alvo (sem percorrer
toda a lista)."""
def contar_menores(lista, alvo):
    inicio, fim = 0, len(lista)
    while inicio < fim:
        meio = (inicio + fim) // 2
        if lista[meio] < alvo:
            inicio = meio + 1
        else:
            fim = meio
    return inicio
print(contar_menores([1, 3, 5, 7], 4))
```

```python
"""busca em matriz ordenada
 Dada uma matriz 4x4 onde cada linha e cada coluna está ordenada, encontre se um número
existe (usando busca binária otimizada)."""
def busca_matriz_ordenada(matriz, alvo):
    if not matriz:
        return (-1, -1)
    
    linhas = len(matriz)
    colunas = len(matriz[0])
    
    linha, coluna = 0, colunas - 1
    
    while linha < linhas and coluna >= 0:
        if matriz[linha][coluna] == alvo:
            return (linha, coluna)
        elif matriz[linha][coluna] > alvo:
            coluna -= 1
        else:
            linha += 1
            
    return (-1, -1)
matriz = [
    [1, 4, 7, 10],
    [2, 5, 8, 11],
    [3, 6, 9, 12],
    [13, 14, 15, 16]
    
```

```python
"""desafio – Rotated Sorted Array
 A lista foi rotacionada (ex: [4,5,6,7,0,1,2]). Encontre o índice do alvo usando busca binária
modificada (sem ordenar primeiro)."""
def busca_rotacionada(lista, alvo):
    inicio, fim = 0, len(lista) - 1
    
    while inicio <= fim:
        meio = (inicio + fim) // 2
        
        if lista[meio] == alvo:
            return meio
        
        if lista[inicio] <= lista[meio]:  # Lado esquerdo ordenado
            if lista[inicio] <= alvo < lista[meio]:
                fim = meio - 1
            else:
                inicio = meio + 1
        else:  # Lado direito ordenado
            if lista[meio] < alvo <= lista[fim]:
                inicio = meio + 1
            else:
                fim = meio - 1
                
    return -1
print(busca_rotacionada([4,5,6,7,0,1,2], 0))
```
